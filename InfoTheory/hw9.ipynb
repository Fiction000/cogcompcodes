```python
awgn = [(p, 1-p) for p in np.linspace(0, 1, 1000)]
```

```python
awgn_gauss = np.random.normal(size=(1000, 2))
```

```python
variances = []
for aw in awgn:
    variances.append(np.var(aw))
```

```python
variances_gauss = []
for aw in awgn_gauss:
    variances_gauss.append(np.var(aw))
```

```python
variances_gauss = np.sort(variances_gauss)
```

```python
snr_dbs_gauss = []
capacities_gauss = []
for var in variances_gauss:
    snr_db = 10 * np.log10(1/var)
    snr_dbs_gauss.append(snr_db)
    capacities_gauss.append(1/2 * np.log2(1 + snr_db))
plt.plot(capacities_gauss, snr_dbs_gauss)
```

```python
snr_dbs = []
capacities = []
for var in variances:
    snr_db = 10 * np.log10(1/var)
    snr_dbs.append(snr_db)
    capacities.append(1/2 * np.log2(1 + snr_db))
plt.plot(capacities, snr_dbs)
```

```python
plt.figure(figsize=(15, 12))
plt.plot(capacities, snr_dbs, label='Binary input distribution')
plt.plot(capacities_gauss, snr_dbs_gauss, label='Gaussian input distribution')
plt.xlabel('Capacity')
plt.ylabel('SNR (db)')
plt.title('The plot of capacity versus SNR in db for binary and gaussian input AWGN channel')
plt.legend()
plt.savefig('capacity_snr.png')
```

```python
plt.figure(figsize=(15, 12))
plt.plot(snr_dbs, capacities, label='Binary input distribution')
plt.plot(snr_dbs_gauss, capacities_gauss, label='Gaussian input distribution')
plt.xlabel('SNR (db)')
plt.ylabel('Capacity')
plt.title('The plot of SNR versus capacity for binary and gaussian input AWGN channel')
plt.legend()
plt.savefig('snr_capacity.png')
```

```python
P = 4
N = 1/4
rhos = []
caps = []

cap = lambda rho: 1/2 * (np.log((N+P) / (N**2 * 2 * np.pi * np.e * (1 - rho**2))))
for i in np.linspace(-0.99999, 0.9999, 1000):
    caps.append(cap(i))
    rhos.append(rhos)
```

```python
plt.plot(rhos, caps)
```

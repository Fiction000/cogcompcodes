```python
import gzip
def generate_cooccurrence_matrix(filepath, word_list=[], sparse=False):
    '''
    input: file path to the cooccurrence file, assuming for each line, the string of the form 'w v COOCCURRENCE_COUNT'
           sparse: if True, create sparse lil matrix instead of raw matrix
    output: cooccurrence word-by-word matrix with (SIZE_OF_VOCABULARY, SIZE_OF_VOCABULARY) 
            dictionary mapping word to index
    '''
    cooccurrence_words = []
    count = 0
    word2idx = {}
    all_words = []

    with gzip.open(filepath, 'rt') as f:
        for line in f:
            if '#(words)' in line: pass
            else:
#                 print(line)
                splitted = line.strip('\n').split()
#                 print(splitted)
                if splitted[0] in word_list or splitted[1] in word_list:
                    all_words.append(splitted[0])
                    all_words.append(splitted[1])
                    cooccurrence_words.append(splitted)

                    count += 1
                    if count % 10000000 == 0:
                        print(count)

    all_words = list(set(all_words))
    
    for word in all_words:
        word2idx[word] = len(word2idx)
        
    idx2word = {idx:word for word,idx in word2idx.items()}
    
    if sparse == True:
        Xc = lil_matrix((len(word2idx), len(word2idx)))
    else: 
        Xc = np.zeros((len(word2idx), len(word2idx))) 
    
    for words in cooccurrence_words:
        i = word2idx[words[0]]
        j = word2idx[words[1]]
        cooccurrence = words[2]
        Xc[i,j] = cooccurrence
        Xc[j,i] = cooccurrence
    
    return Xc, word2idx, idx2word
```
